```python
#4 - O volume de um cubo é determinado através do produto da área da base pela altura,
#    (mas as arestas do cubo possuem medidas iguais), então temos que:
#    V = Ab * a ou V = a * a * a → V = a³. A partir disso, faça um programa, adequando as variáveis
#    para receber medidas de uma piscina (altura, largura e comprimento), para responder o volume de
#    água necessário para enchê-la.

def calcular_volume_piscina():
  try:
    altura = float(input("Digite a altura da piscina (em metros, ex: 5): "))
    largura = float(input("Digite a largura da piscina (em metros, ex: 3): "))
    comprimento = float(input("Digite o comprimento da piscina (em metros, ex: 4): "))
    if altura <= 0 or largura <= 0 or comprimento <= 0:
      print("Erro: Todas as medidas devem ser maiores que zero.")
      return
    volume = altura * largura * comprimento
    print(f"O volume da piscina é {volume:.2f} m³.")
  except ValueError:
    print("Por favor, digite um número válido para todas as entradas.")
```

```python
#6 - Construa um programa em Python em que o usuário insira a sigla de um estado brasileiro em que
#    uma pessoa nasceu e, em seguida imprima uma das seguintes mensagens:
#    Carioca Paulista Mineiro Outros estados

def verificar_estado_nascimento():
  sigla_estado = input("Digite a sigla de um estado brasileiro: ")
  sigla_estado = sigla_estado.upper()
  if sigla_estado == "RJ":
    print("Carioca")
  if sigla_estado == "SP":
    print("Paulista")
  if sigla_estado == "MG":
    print("Mineiro")
  if sigla_estado == "RS":
    print("Gaucho")
  else:
    print("Outros estados")

verificar_estado_nascimento()
```

```python
#7 - Faça um programa Python que receba duas notas, calcule a média aritmética e mostre o resultado.
#    Além disso, deve mostrar ao lado da média Aprovado (se média >= 7.0) Reprovado (se média <= 3.0),
#    Exame (se média entre 3.0 e 7.0)

notas = []

def menu_medias():
  print("Escolha uma opção:")
  print("1 - Calcular média de notas")
  print("2 - Inserir nota")
  print("3 - Sair")
  opcao = input("Digite o número da opção desejada: ")
  if opcao == "1":
    calcular_media_notas()
  if opcao == "2":
    inserir_nota()
  if opcao == "3":
    print("Saindo do programa...")
    return

def inserir_nota():
  try:
    nova_nota = float(input("Digite a nova nota (entre 0 e 10): "))
    if 0 <= nova_nota <= 10:
      notas.append(nova_nota)
      print(f"Nota {nova_nota} inserida com sucesso!")
      menu_medias()
  except ValueError:
    print("Erro: Por favor, insira uma nota válida (entre 0 e 10).")
  except Exception as e:
    print(f"Ocorreu um erro inesperado: {e}")

def calcular_media_notas():

  if not notas:
      print("Erro: Nenhuma nota foi inserida.")
      menu_medias()

  soma_notas = sum(notas)
  quantidade_notas = len(notas)
  media = soma_notas / quantidade_notas

  print(f"A média das notas é: {media:.2f}")

  if media >= 7.0:
    print(f"Aprovado! Média: {media:.2f}")
  elif media <= 3.0:
    print(f"Reprovado! Média: {media:.2f}")
  else:
    print(f"Exame! Média: {media:.2f}")

  menu_medias()

menu_medias()
```

```python
def calcular_insulina():
  try:
    glicemia_do_momento = float(input("Digite sua glicemia do momento (mg/dl, ex: 250): "))
    meta_pre_refeicao = float(input("Digite sua meta pré-refeição (mg/dl, ex: 100): "))
    fator_sensibilidade = float(input("Digite seu fator de sensibilidade (valor entre 20 e 60, ex: 40): "))

    if not (20 <= fator_sensibilidade <= 60):
      print("Erro: O fator de sensibilidade deve estar entre 20 e 60.")
      return

    quantidade_insulina = (glicemia_do_momento - meta_pre_refeicao) / fator_sensibilidade

    print(f"Você deverá administrar {quantidade_insulina:.2f} unidades de insulina.")

  except ValueError:
    print("Por favor, digite um número válido para todas as entradas.")
  except ZeroDivisionError:
    print("Erro: Divisão por zero. Verifique o fator de sensibilidade.")
  except Exception as e:
    print(f"Ocorreu um erro inesperado: {e}")

calcular_insulina()
```
