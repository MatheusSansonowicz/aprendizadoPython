## Inicializar o Sistema GPS e a Rede de Cidades

### Subtask:
Definir a classe `GPSNetwork` para gerenciar as conexões entre cidades e instanciar a rede. Esta classe terá métodos para adicionar/atualizar rotas, obter distâncias e exibir a rede.


**Reasoning**:
To address the subtask's instructions, I will define the `GPSNetwork` class with its methods (`__init__`, `adicionar_ou_atualizar_rota`, `obter_distancia`, `exibir_rede`) and then instantiate an object of this class. This aligns with the request to define the class and instantiate the network.



```python
class GPSNetwork:
  def __init__(self):
    self.rotas = {}

  def adicionar_ou_atualizar_rota(self, origem, destino, distancia):
    if origem not in self.rotas:
      self.rotas[origem] = {}
    self.rotas[origem][destino] = distancia

    if destino not in self.rotas:
      self.rotas[destino] = {}
    self.rotas[destino][origem] = distancia
    print(f"Rota entre {origem} e {destino} ({distancia} km) adicionada/atualizada.")

  def obter_distancia(self, origem, destino):
    if origem in self.rotas and destino in self.rotas[origem]:
      return self.rotas[origem][destino]
    return None

  def exibir_rede(self):
    print("\n--- Rede GPS ---")
    if not self.rotas:
      print("A rede está vazia.")
      return

    for origem, destinos in self.rotas.items():
      print(f"Cidade: {origem}")
      for destino, distancia in destinos.items():
        print(f"  -> {destino}: {distancia} km")
    print("----------------")

# Criar uma instância da classe GPSNetwork
rede_gps = GPSNetwork()
print("Instância da rede GPS criada com sucesso.")
```
