```python
# 1. O Reflorestador (Loop com Acumulador)
# Em um mini-game de reflorestamento, o jogador deve plantar árvores até atingir
# uma meta de "Biomassa". O programa deve parar assim que a meta for atingida
# ou superada e exibir quantas árvores foram necessárias.

try:
    meta_biomassa = float(input('Digite a meta de biomassa desejada: '))
    biomassa_atual = 0.0
    quantidade_arvores = 0

    while biomassa_atual < meta_biomassa:
        try:
            entrada = input(f'Digite a biomassa da árvore {quantidade_arvores + 1}: ')
            valor_arvore = float(entrada)

            biomassa_atual += valor_arvore
            quantidade_arvores += 1
        except ValueError:
            print('Erro: Por favor, digite um valor numérico válido para a biomassa.')
            continue

    print(f'\nMeta atingida!')
    print(f'Total de biomassa: {biomassa_atual:.2f}')
    print(f'Quantidade de árvores necessárias: {quantidade_arvores}')

except ValueError:
    print('Erro: A meta de biomassa deve ser um número.')
```

```python
#2. Simulador de Combate: Turnos de RPG
#Em um jogo de turno, um herói tem 100 de HP e enfrenta
# um monstro com 100 de HP. O programa deve pedir ao usuário
# o dano do ataque do herói e o dano do ataque do monstro por rodada.
# Use um while que continue enquanto ambos estiverem
#com HP acima de zero. Ao final de cada turno, mostre o HP restante de ambos.
# Se alguém chegar a 0, encerre e anuncie o vencedor.

vida_her0i = 100
vida_monstro = 100

while vida_her0i > 0  or vida_monstro > 0:
    try:
        print('Bem vindo ao jogo de RPG!')
        print(f'Vida do Herói: {vida_her0i}')
        print(f'Vida do Monstro: {vida_monstro}')
        entrada = input('Digite o dano do ataque do herói: ')
        dano_her0i = float(entrada)
        vida_monstro -= dano_her0i
        entrada = input('Digite o dano do ataque do monstro: ')
        dano_monstro = float(entrada)
        vida_her0i -= dano_monstro

    except ValueError:
        print('Erro: Por favor, digite um valor valido.')
        continue

print(f'\nBatalha decidida!')
if vida_her0i <= 0:
    print('O Monstro venceu!')
if vida_monstro <= 0:
    print('O Herói venceu!')
```

```python
#5. Média de Temperatura Global
#Para um painel informativo sobre mudanças climáticas,
# você precisa calcular a média de temperatura de um período.
# O usuário deve informar quantos dias deseja analisar.
# Use o while para ler a temperatura de cada um desses dias.
# Ao final, o programa deve exibir a média aritmética das temperaturas
# e informar se a média está "Acima do esperado" (caso seja maior que 25°C)
# ou "Dentro da normalidade".

try:
    quantidade_dias = int(input('Quantos dias deseja analisar? '))

    if quantidade_dias <= 0:
        print('Por favor, insira um número de dias maior que zero.')
    else:
        soma_temperaturas = 0.0
        contador = 0

        while contador < quantidade_dias:
            entrada = input(f'Digite a temperatura do dia {contador + 1}: ')
            try:
                temp = float(entrada)
                soma_temperaturas += temp
                contador += 1
            except ValueError:
                print('Erro: Digite um valor numérico válido para a temperatura.')

        media = soma_temperaturas / quantidade_dias

        print(f'\nMédia aritmética das temperaturas: {media:.2f}°C')

        if media > 25:
            print('Status: Acima do esperado')
        else:
            print('Status: Dentro da normalidade')

except ValueError:
    print('Erro: A quantidade de dias deve ser um número inteiro.')
```
